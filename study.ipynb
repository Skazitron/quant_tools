```python
import numpy as np
from numpy import linalg as la

A = np.array([
    [2,1],
    [1, 2]
])

w, v = la.eig(A)
```

```python
A_10 = v @ np.diag(w**10) @ la.inv(v)
A_10
```

```python
B = np.array([
    [1, 1],
    [0, 2]]
)

b_eigenvalues, b_eigenvectors = la.eig(B)
np.sum(b_eigenvectors@(np.diag(b_eigenvalues**10))@la.inv(b_eigenvectors))
```
