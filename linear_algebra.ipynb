```python
def check_orthogonality(matrix: np.ndarray) -> bool:
    e_v, matrix = la.eig(matrix) 
    product = matrix @ matrix.T
    identity = np.eye(matrix.shape[0])
    return np.allclose(product, identity)

check_orthogonality(np.array([
    [2,0],
    [0,2]
]))
```
